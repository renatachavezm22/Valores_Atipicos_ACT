### **Limpieza de los datos nulos**

CONSIDERANDO QUE REQUIERO EL MENOR SESGO DE LA INFORMACIÓN ES QUE TRATARE DE ELIMINAR EL MENOR NÚMERO DE COLUMNAS Y EL DE POLIZA ASÍ COMO EL DE FOLIO SE LE ASIGNARA UN VALOR EN TODOS LOS DATAFRAME.

```python
gastos2020SN=gastos2020.copy()
```

```python
gastos2020SN=gastos2020.drop('GASTO', axis=1)
gastos2020SN
```

```python
# Eliminar los nulos
gastos2020SN=gastos2020SN.fillna(round(gastos2020.mean(),1))
gastos2020SN['POLIZA']=gastos2020SN['POLIZA'].fillna('SIN POLIZA')
gastos2020SN['FOLIO']=gastos2020SN['FOLIO'].fillna(000000)
gastos2020SN['TIPO']=gastos2020SN['TIPO'].fillna(method='bfill')

gastos2020SN
```

```python
vn1=gastos2020SN.isnull().sum()
vn1
```

```python
gastos2021SN=gastos2021.copy()
```

```python
vn2=gastos2021SN.isnull().sum()
vn2
```

```python
gastos2022SN=gastos2022.copy()
gastos2022SN=gastos2022SN.drop('Otros', axis=1)
```

```python
gastos2022SN['Poliza']=gastos2022SN['Poliza'].fillna('SIN POLIZA')
gastos2022SN['Folio']=gastos2022SN['Folio'].fillna(000000)
gastos2022SN=gastos2022SN.fillna(method='ffill')
gastos2022SN
```

UNA VEZ QUE ELIMINAMOS LAS VARIABLES DE TIPO STRING ES QUE PODEMOS PROCEDES A VER LOS OUTLIERS Y EL PROCEDIMIENTO DE INTERCUARTIL

### **Gastos 2020 Outliers e Intercuatil**



```python
fig1 = plt.figure(figsize =(15, 8))
gastos2020SN1.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe GASTOS 2020SN (YA LIMPIO)")
plt.show() #dibujamos el diagrama
```

```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y1=gastos2020SN1
Limite_Superior= y1.mean() + 3*y1.std()
Limite_Inferior= y1.mean() - 3*y1.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos Ouliers
outliers1= gastos2020SN1[(y1>Limite_Superior)|(y1<Limite_Inferior)]
outliers1
```

```python
gastos2020_clean= gastos2020SN1[(y1<=Limite_Superior)&(y1>Limite_Inferior)]
gastos2020_clean
```

```python
fig1 = plt.figure(figsize =(15, 8))
gastos2020_clean.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe GASTOS 2020SN (YA LIMPIO)")
plt.show() #dibujamos el diagrama
```

Intercuartilico

```python
y1=gastos2020SN1

percentile25=y1.quantile(0.25) #Q1
percentile75=y1.quantile(0.75) #Q3
iqr1= percentile75 - percentile25

Limite_Superior_iqr1= percentile75 + 1.5*iqr1
Limite_Inferior_iqr1= percentile25 - 1.5*iqr1
print("Limite superior permitido", Limite_Superior_iqr1)
print("Limite inferior permitido", Limite_Inferior_iqr1)
```

```python
#Encontramos Ouliers
outliers_iqr1= gastos2020SN1[(y1>Limite_Superior_iqr1)|(y1<Limite_Inferior_iqr1)]
outliers_iqr1
```

```python
#Obtenemos datos limpios
gastos2020_clean_iqr= gastos2020SN1[(y1<=Limite_Superior_iqr1)&(y1>=Limite_Inferior_iqr1)]
gastos2020_clean_iqr
```

```python
fig2 = plt.figure(figsize =(15, 8))
gastos2020_clean_iqr.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe GASTOS 2020SN (YA LIMPIO)")
plt.show() #dibujamos el diagrama
```

Como podemos ver tenemos un mejor desempeño en el método intercuartilico considerando que me reduce el rango de outliers de 350000 a 35000 en los de la desviación estandar y 8000 intercuartilico
